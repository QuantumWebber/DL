```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("DailyDelhiClimate.csv")
```

```python
df=df[["date","meantemp"]]
df.head()
```

```python
x1 = df['meantemp'].ewm(alpha=0.9).mean()
```

```python
x1
```

```python
df['ewma']=x1
df
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
plt.plot(df['date'],x1,color='black',label='alpha=0.9')

plt.legend()
plt.show
```
