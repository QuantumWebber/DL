<h4>Memoization is a optimization technique jisme hum computer program ko speed up krne ke liye result store krlete ek function ka taki usse baar baar code na krna pade </h4>

```python
def fib(n):
    if n==1 or n==0:
        return 1
    else:
        return fib(n-1)+fib(n-2)
```

```python
fib(10)
```

```python
# this code is very ineffective it takes too much time for larger number
```

```python
#how memoization solve this problem of exponential time complexity problem 
```

```python
def fibi(n, d):
    if n in d:
        return d[n]
    else:
        d[n] = fibi(n - 1, d) + fibi(n - 2, d)
        return d[n]
```

```python
memo = {0: 0, 1: 1}
print(fibi(38, memo))
```

```python
# in short memoization space ka kharcha krke time ko reduce krta 
```
